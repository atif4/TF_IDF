```python
# EDA pamages 
import pandas as pd

#hide wearrings 
import warnings

warnings.filterwarnings('ignore')
#import the necessary module

from sklearn import preprocessing
# create the Labelencoder object
le = preprocessing.LabelEncoder()
#convert the categorical columns into numeric
le = preprocessing.LabelEncoder()

from sklearn.model_selection import train_test_split
#split data set into train and test sets

# import the necessary module
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
#create an object of the type GaussianNB
gnb = GaussianNB()

#import necessary modules
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
class Dataset_tf():
    """A simple attempt to model a dog."""
    def __init__(self, df):
        """Initialize name and age attributes."""
        self.df = pd.read_csv(df)
        
    def get_convert_dtype(self,list_of_c_ids):
        for list_of_c_id in list_of_c_ids:
            self.df[list_of_c_id] = self.df[list_of_c_id].apply(pd.to_numeric)
            # convert just columns "a" and "b" df[["a", "b"]] = df[["a", "b"]].apply(pd.to_numeric)
            self.df[list_of_c_id] = self.df[list_of_c_id].astype(int)
        
    def get_LabelEncoder(self,labelencoders_list):
        for labelencoder_list in labelencoders_list:
            self.df[labelencoder_list] = le.fit_transform(self.df[labelencoder_list])
            #return self.df[labelencoder_list]
    def get_divided_df(self,c_id1,c_id2):
        # select columns other than 'Opportunity Number','Opportunity Result'
        cols = [col for col in self.df.columns if col not in [c_id1,c_id2]]
        global data , target
        # dropping the 'Opportunity Number'and 'Opportunity Result' columns
        data = self.df[cols]
        #assigning the Oppurtunity Result column as target
        target = self.df[c_id1]
        return data
    def get_test_train(self):
        data_train, data_test, target_train, target_test = train_test_split(data  , target , test_size = 0.30, random_state = 10)
        return data_train, data_test, target_train, target_test
    def get_prediction(self):
        #train the algorithm on training data and predict using the testing data
        pred = gnb.fit(data_train, target_train).predict(data_test)
        print('Prediction',pred.tolist())
        print ('the accuracy score of the model')
        print("Naive-Bayes accuracy : ",accuracy_score(target_test, pred, normalize = True))
    def get_lassifier(self):
        #create object of the lassifier
        neigh = KNeighborsClassifier(n_neighbors=3)
        #Train the algorithm
        neigh.fit(data_train, target_train)
        # predict the response
        pred = neigh.predict(data_test)
        # evaluate accuracy
        print ("KNeighbors accuracy score : ",accuracy_score(target_test, pred))
```
